```python
datasets = [dataset_1, dataset_2, dataset_3, dataset_4, dataset_5]
```

```python
structured_features = ['age', 'sex', 'heart_rate', 'bp_systolic', 'bp_diastolic', 'resp_rate', 'temperature_C', 'oxygen_saturation','ESI', 'recent_admissions_30d']
text_feature = 'triage_notes'
response_var = 'admitted'

all_targets = set(structured_features + [text_feature, response_var])
```

```python
col_mapping = {
    # age
    'age': 'age',

    # sex
    'sex': 'sex',
    'gender': 'sex',

    # heart rate
    'heart_rate': 'heart_rate',
    'hr': 'heart_rate',

    # blood pressure
    'bp_systolic': 'bp_systolic',
    'systolic_bp': 'bp_systolic',

    'bp_diastolic': 'bp_diastolic',
    'diastolic_bp': 'bp_diastolic',

    # respiratory rate
    'resp_rate': 'resp_rate',
    'respiratory_rate': 'resp_rate',

    # temperature
    'temperature_c': 'temperature_C',
    'temperature': 'temperature_C',

    # oxygen saturation
    'oxygen_saturation': 'oxygen_saturation',
    'o2_sat': 'oxygen_saturation',
    'oxygen_sat': 'oxygen_saturation',

    # ESI
    'esi': 'ESI',
    'esi_score': 'ESI',

    # recent admissions
    'recent_admissions_30d': 'recent_admissions_30d',
    'recent_admissions': 'recent_admissions_30d',

    # triage notes
    'triage_notes': 'triage_notes',
    'triage_note': 'triage_notes',

    # admitted
    'admitted': 'admitted'
}
```

```python
def standardize_df(df: pd.DataFrame) -> pd.DataFrame:
    rename_dict = {}
    for col in df.columns:
        key = col.strip().lower()
        if key in col_mapping:
            rename_dict[col] = col_mapping[key]

    # rename only the recognized ones
    df = df.rename(columns=rename_dict)

    # now keep *only* the columns in our target set
    keep = [c for c in df.columns if c in all_targets]
    return df[keep]
```

```python
cleaned = []

for dataset in datasets:
  df_std = standardize_df(dataset)
  print(f"{dataset}: -> kept: {df_std.columns.tolist()}")
  cleaned.append(df_std)
```

```python
merged_df = pd.concat(cleaned, ignore_index=True)
print("\nMerged shape:", merged_df.shape)
print("Columns:", merged_df.columns.tolist())
```
