```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Updated_merged_data.csv')

data.head()
```

```python
structured_features = ['age', 'sex', 'heart_rate', 'bp_systolic', 'bp_diastolic', 'resp_rate', 'temperature_C', 'oxygen_saturation', 'ESI', 'recent_admissions_30d']
text_feature = 'triage_notes'
response_var = 'admitted'
```

```python
def create_blended_notes(row):
    structured_info = ""
    for feature in structured_features:
        value = row[feature]
        if pd.notna(value):
          structured_info += f"{feature}: {value} / "

    blended_notes = f"{structured_info} {row[text_feature]}"
    return blended_notes

data['blended_notes'] = data.apply(create_blended_notes, axis=1)

data[['blended_notes', 'triage_notes']].head()
```

```python
blended_data = data[['blended_notes', 'admitted']]

blended_data.head()
```

```python
blended_data.to_csv('Updated_blended_data.csv', index=False)
```
